## 4. Análisis de mercado.

### Número de libros publicados después del 1 de enero del 2000.

```python
query = """
SELECT COUNT(book_id)
FROM books
WHERE publication_date > '2000-01-01'
"""
result = pd.io.sql.read_sql(query, con = engine)
print('Número de libros:')
print(result)
```

Después de verificar cuántos libros hay después del 2001, podemos ver que mas del 80% de estos fueron publicados despues de esta fecha.

### Número de reseñas de usuarios y la calificación promedio de cada libro.

```python
query = """ 
SELECT title, 
       COUNT(review_id) AS total_reseñas, 
       AVG(rating) AS calificacion_promedio
FROM books 
JOIN reviews ON books.book_id = reviews.book_id
JOIN ratings ON books.book_id = ratings.book_id
GROUP BY title, books.book_id
"""
result_cal = pd.io.sql.read_sql(query, con = engine)
print('Total de reseñas y sus promedios:\n')
print(result_cal)
```

Después de verificar cuántos libros tienen reseñas, podemos ver una gran participación en esto; prácticamente todos los libros tienen reseñas, aunque entre estos hay una gran varianza de entre 4 y 60 reseñas, y la calificación promedio de estos es bastante alta, lo cual también puede indicar una gran aceptación de los usuarios a los libros publicados.

### Nombres de editoriales que tienen más libros con más de 50 páginas.

```python
query = """
SELECT publisher, COUNT(*)
FROM books 
JOIN publishers ON books.publisher_id = publishers.publisher_id
WHERE num_pages > 50 
GROUP BY publishers.publisher
ORDER BY COUNT(*) DESC
"""

result_50 = pd.io.sql.read_sql(query, con = engine)
print('Editoriales que tienen mayor numero de libros con mas de 50 paginas:\n')
print(result_50)
```

Después de verificar entre las editoriales (334) y aplicarles un filtro para que solo se calificaran a editoriales en que sus libros tengan más de 50 páginas, podemos ver que Penguin Books es la editorial con la mayor cantidad de estos, con 42 libros, seguido por Vintage con 31 y Grand Central Publishing con 25. 

## Autor que tiene la más alta calificación (al menos 50 calificaciones registradas).

```python
query = """
SELECT author AS autor, AVG(rating) AS promedio
FROM books
JOIN authors ON books.author_id = authors.author_id
JOIN ratings ON books.book_id = ratings.book_id
GROUP BY author
HAVING COUNT(rating) > 50
"""

result_author = pd.io.sql.read_sql(query, con = engine)
print('Promedio de las calificaciones de los autores con 50 calificaciones:\n')
print(result_author)
```

Podemos ver que solo 22 autores cuentan con más de 50 calificaciones en sus libros y estos cuentan con calificaciones más altas de la media. Destacando, J.K. Rowling tiene la calificación más alta (4.29). Agatha Christie también está muy bien valorada (4.28). J. R. R. Tolkien tiene una excelente calificación (4.24), pero notemos que John Steinbeck tiene la calificación más baja del grupo (3.64). Esto nos da una idea de la aceptación de los usuarios, la cual podemos concluir que es buena.

## Número promedio de reseñas entre los usuarios que calificaron mas de 50 libros.

```python
query = """
SELECT reviews.username, 
    COUNT(reviews.book_id) AS numere_de_calificaciones, 
    AVG(rating) AS promedio
FROM ratings
JOIN reviews ON ratings.book_id = reviews.book_id 
GROUP BY reviews.username
HAVING COUNT(reviews.book_id) > 50
"""
result_user = pd.io.sql.read_sql(query, con = engine)
print('Calificacion promedio de usuarios que hicieron mas de 50 reseñas:\n')
print(result_user)
```

Podemos ver que los usuarios dejan bastantes reseñas en los libros que leen y el promedio de las calificaciones es bastante alto para la mayoría de los libros, aunque podemos decir que no hay un patrón a seguir en esta consulta, ya que hay usuarios con muchas reseñas y baja calificación promedio, y lo mismo pasa por el contrario, pero podemos decir que los usuarios aceptan positivamente los libros que se publican en la plataforma.
