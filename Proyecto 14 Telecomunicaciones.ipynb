### Decisión tomada: ELIMINAR DUPLICADOS
**Justificación:**
1. **Objetivo del negocio:** Identificar operadores ineficaces requiere métricas precisas
2. **No aportan información nueva:** Son registros idénticos
3. **Sesgo en métricas:** Inflan artificialmente el volumen de llamadas por operador
4. **Mantienen representatividad:** Las características se conservan

### Estrategia de implementación:
- Eliminar duplicados usando `drop_duplicates()`
- Comparar métricas antes/después para validar consistencia
- Documentar el impacto en el análisis

```python
print("ELIMINACIÓN DE DUPLICADOS")

print("ANTES de eliminar duplicados:")
print(f"Total de registros: {len(telecom_data)}")
print(f"Llamadas perdidas: {telecom_data['is_missed_call'].mean()*100:.1f}%")
print(f"Operadores únicos: {telecom_data['operator_id'].nunique()}")

telecom_data_clean = telecom_data.drop_duplicates()

print(f"\nDESPUÉS de eliminar duplicados:")
print(f"Total de registros: {len(telecom_data_clean)}")
print(f"Registros eliminados: {len(telecom_data) - len(telecom_data_clean)}")
print(f"Llamadas perdidas: {telecom_data_clean['is_missed_call'].mean()*100:.1f}%")
print(f"Operadores únicos: {telecom_data_clean['operator_id'].nunique()}")

telecom_data = telecom_data_clean.copy()
print(f"\n Dataset actualizado con {len(telecom_data)} registros limpios")
```

```python
print("VALIDACIÓN DEL IMPACTO EN ANÁLISIS DE OPERADORES")

telecom_operators_clean = telecom_data.dropna(subset=['operator_id']).copy()

operator_performance_clean = telecom_operators_clean.groupby('operator_id').agg({
    'is_missed_call': ['count', 'sum', 'mean'],
    'call_duration': 'mean',
    'calls_count': 'sum'
}).round(3)

operator_performance_clean.columns = ['total_llamadas', 'llamadas_perdidas', 'tasa_perdidas',
                                    'duracion_promedio', 'volumen_total']

operator_performance_clean = operator_performance_clean[operator_performance_clean['total_llamadas'] >= 100]

print(f"Operadores analizados después de limpieza: {len(operator_performance_clean)}")
print("\nTop 5 operadores más ineficaces (datos limpios):")
print(operator_performance_clean.sort_values('tasa_perdidas', ascending=False).head())

print(f"\n Análisis de operadores actualizado con datos limpios")
print(f" Ahora tienes {len(operator_performance_clean)} operadores para análisis confiable")
```

#### Análisis de patrones temporales.

```python
telecom_data['year'] = telecom_data['date'].dt.year
telecom_data['month'] = telecom_data['date'].dt.month
telecom_data['day'] = telecom_data['date'].dt.day
telecom_data['weekday'] = telecom_data['date'].dt.dayofweek  
telecom_data['hour'] = telecom_data['date'].dt.hour
telecom_data['week'] = telecom_data['date'].dt.isocalendar().week

print("Variables temporales creadas exitosamente!")
print(f"Rango de fechas: {telecom_data['date'].min()} a {telecom_data['date'].max()}")
```

```python
print("ANÁLISIS TEMPORAL\n")

print("1. Distribución por mes:")
monthly_calls = telecom_data.groupby('month').agg({
    'calls_count': 'sum',
    'call_duration': 'sum',
    'is_missed_call': 'mean'
}).round(2)
print(monthly_calls)

print("\n2. Distribución por día de la semana:")
days_names = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
weekly_calls = telecom_data.groupby('weekday').agg({
    'calls_count': 'sum',
    'call_duration': 'sum',
    'is_missed_call': 'mean'
}).round(2)
weekly_calls.index = days_names
print(weekly_calls)
```

Vemos un crecimiento progresivo mes con mes hasta el mes de noviembre, que se mantiene en un nivel alto.

Por día de la semana, los días laborales suelen ser más activos que los fines de semana y esto se refleja muy bien con el porcentaje tan alto de llamadas perdidas (51%).



```python
print("3. Distribución por hora del día:\n")
hourly_calls = telecom_data.groupby('hour').agg({
    'calls_count': 'sum',
    'call_duration': 'mean',
    'is_missed_call': 'mean'
}).round(2)

print("Horas con más actividad:")
print(hourly_calls.sort_values('calls_count', ascending=False).head(5))

print("\nHoras con mayor duración promedio:")
print(hourly_calls.sort_values('call_duration', ascending=False).head(5))
```

En esta parte podemos notar que la hora con más actividad y la que tiene más duración es la misma, pero este dato no nos es relevante, ya que todas las entradas tienen la misma hora.

## FORMULACIÓN DE HIPÓTESIS
Hipótesis nula: La tasa de llamadas perdidas del operador más ineficaz es igual a la tasa general de la empresa

Hipótesis alternativa: La tasa de llamadas perdidas del operador más ineficaz es significativamente mayor que la tasa general

Tipo de prueba: Prueba Z para una proporción (cola derecha)

Nivel de significancia: α = 0.05

```python
tasa_general = telecom_data['is_missed_call'].mean()
print(f"\nTasa general de llamadas perdidas: {tasa_general:.3f} ({tasa_general*100:.1f}%)")

operador_mas_ineficaz = operator_performance_clean.sort_values('tasa_perdidas', ascending=False).index[0]
tasa_operador_ineficaz = operator_performance_clean.loc[operador_mas_ineficaz, 'tasa_perdidas']

print(f"Operador más ineficaz: {operador_mas_ineficaz}")
print(f"Tasa del operador más ineficaz: {tasa_operador_ineficaz:.3f} ({tasa_operador_ineficaz*100:.1f}%)")
```

```python
print("PRUEBA ESTADÍSTICA: Z-TEST PARA UNA PROPORCIÓN ")

operador_data = telecom_operators[telecom_operators['operator_id'] == operador_mas_ineficaz]
n_operador = len(operador_data)  
x_operador = operador_data['is_missed_call'].sum() 
p_operador = x_operador / n_operador  

p_general = tasa_general  

print(f"DATOS DE LA PRUEBA:")
print(f"Operador {operador_mas_ineficaz}:")
print(f"  - Llamadas totales: {n_operador}")
print(f"  - Llamadas perdidas: {x_operador}")
print(f"  - Proporción: {p_operador:.3f}")
print(f"Proporción general: {p_general:.3f}")

print(f"\nVERIFICACIÓN DE CONDICIONES:")
print(f"np = {n_operador * p_general:.1f} ≥ 5: {'✅' if n_operador * p_general >= 5 else '❌'}")
print(f"n(1-p) = {n_operador * (1-p_general):.1f} ≥ 5: {'✅' if n_operador * (1-p_general) >= 5 else '❌'}")

z_score = (p_operador - p_general) / np.sqrt(p_general * (1 - p_general) / n_operador)
p_value = 1 - stats.norm.cdf(z_score)  # cola derecha

print(f"\nRESULTADOS DE LA PRUEBA:")
print(f"Estadístico Z: {z_score:.4f}")
print(f"P-valor: {p_value:.6f}")
print(f"Nivel de significancia: α = 0.05")

if p_value < 0.05:
    print(f"\n CONCLUSIÓN: RECHAZAMOS H0")
    print(f"   Evidencia estadísticamente significativa de que el operador {operador_mas_ineficaz}")
    print(f"   tiene una tasa de llamadas perdidas significativamente MAYOR que el promedio general")
    print(f"   (p-valor = {p_value:.6f} < 0.05)")
else:
    print(f"\nCONCLUSIÓN: NO RECHAZAMOS H0")
    print(f"   No hay evidencia estadísticamente significativa de diferencia")
    print(f"   (p-valor = {p_value:.6f} ≥ 0.05)")
```

```python
print("ANÁLISIS ESTADÍSTICO DE MÚLTIPLES OPERADORES")

top_5_ineficaces = operator_performance_clean.sort_values('tasa_perdidas', ascending=False).head(5)

print("PRUEBAS INDIVIDUALES PARA TOP 5 OPERADORES INEFICACES:")
print("-" * 60)

resultados_pruebas = []

for operador_id in top_5_ineficaces.index:
    
    operador_data = telecom_operators_clean[telecom_operators_clean['operator_id'] == operador_id]
    n_op = len(operador_data)
    x_op = operador_data['is_missed_call'].sum()
    p_op = x_op / n_op
    
    z_score = (p_op - tasa_general) / np.sqrt(tasa_general * (1 - tasa_general) / n_op)
    p_value = 1 - stats.norm.cdf(z_score)
    
    significativo = "SÍ" if p_value < 0.05 else "NO"
    
    print(f"Operador {operador_id}:")
    print(f"  Tasa: {p_op:.3f} vs General: {tasa_general:.3f}")
    print(f"  Z-score: {z_score:.4f} | P-valor: {p_value:.6f}")
    print(f"  Significativo (α=0.05): {significativo}")
    print()
    
    # Guardar resultados
    resultados_pruebas.append({
        'operador_id': operador_id,
        'tasa_operador': p_op,
        'z_score': z_score,
        'p_value': p_value,
        'significativo': p_value < 0.05
    })

operadores_significativos = sum(1 for r in resultados_pruebas if r['significativo'])
print(f"📊 RESUMEN:")
print(f"   Operadores analizados: {len(resultados_pruebas)}")
print(f"   Operadores estadísticamente ineficaces: {operadores_significativos}")
print(f"   Porcentaje de confirmación: {operadores_significativos/len(resultados_pruebas)*100:.1f}%")
```

```python
print("CORRECCIÓN POR MÚLTIPLES COMPARACIONES (BONFERRONI)")

alpha_original = 0.05
num_pruebas = len(resultados_pruebas)
alpha_corregido = alpha_original / num_pruebas

print(f"Nivel de significancia original: α = {alpha_original}")
print(f"Número de pruebas: {num_pruebas}")
print(f"Nivel de significancia corregido: α = {alpha_corregido:.4f}")
print()

print("RESULTADOS CON CORRECCIÓN DE BONFERRONI:")
print("-" * 50)

operadores_significativos_corregido = 0
for resultado in resultados_pruebas:
    significativo_corregido = resultado['p_value'] < alpha_corregido
    if significativo_corregido:
        operadores_significativos_corregido += 1

    print(f"Operador {resultado['operador_id']}:")
    print(f"  P-valor: {resultado['p_value']:.6f}")
    print(f"  α corregido: {alpha_corregido:.4f}")
    print(f"  Significativo (Bonferroni): {'SÍ' if significativo_corregido else 'NO'}")
    print()

print(f"📊 RESUMEN CON CORRECCIÓN DE BONFERRONI:")
print(f"   Operadores significativos (α original): {sum(1 for r in resultados_pruebas if r['p_value'] < 0.05)}")
print(f"   Operadores significativos (α corregido): {operadores_significativos_corregido}")
print(f"   Impacto de la corrección: {'Reduce falsos positivos' if operadores_significativos_corregido < sum(1 for r in resultados_pruebas if r['p_value'] < 0.05) else 'Sin cambios'}")

```

```python
print("=== CONCLUSIONES DE LA INFERENCIA ESTADÍSTICA ===")
print()
df_resultados = pd.DataFrame(resultados_pruebas)
df_resultados['significativo_bonferroni'] = df_resultados['p_value'] < alpha_corregido

print("📋 TABLA RESUMEN DE RESULTADOS:")
print("-" * 80)
for _, row in df_resultados.iterrows():
    print(f"Operador {row['operador_id']}:")
    print(f"  Tasa de llamadas perdidas: {row['tasa_operador']:.3f} ({row['tasa_operador']*100:.1f}%)")
    print(f"  Diferencia vs promedio: +{(row['tasa_operador'] - tasa_general)*100:.1f} puntos porcentuales")
    print(f"  Z-score: {row['z_score']:.3f}")
    print(f"  P-valor: {row['p_value']:.6f}")
    print(f"  Significativo (α=0.05): {'✅ SÍ' if row['significativo'] else '❌ NO'}")
    print(f"  Significativo (Bonferroni): {'✅ SÍ' if row['significativo_bonferroni'] else '❌ NO'}")
```

```python
print(" CONCLUSIONES FINALES:")
print("-" * 40)
operadores_confirmados = df_resultados['significativo_bonferroni'].sum()
if operadores_confirmados > 0:
    print(f"✅ Se confirma estadísticamente que {operadores_confirmados} operador(es)")
    print(f"   tienen tasas de llamadas perdidas significativamente MAYORES al promedio")
    print(f"   con un nivel de confianza del 95% (después de corrección por múltiples comparaciones)")
else:
    print("❌ Después de la corrección de Bonferroni, no se puede confirmar")
    print("   estadísticamente que algún operador sea significativamente más ineficaz")

print(f"\nIMPACTO EMPRESARIAL:")
if operadores_confirmados > 0:
    operadores_problema = df_resultados[df_resultados['significativo_bonferroni']]['operador_id'].tolist()
    print(f"   Operadores que requieren intervención inmediata: {operadores_problema}")
    print(f"   Estos operadores representan un problema estadísticamente comprobado")
else:
    print("   Las diferencias observadas podrían deberse a variabilidad natural")
    print("   Se recomienda análisis adicional antes de tomar acciones correctivas")
```

Despues de realizar multiples pruebas podemos concluir que nuestro peor operador tiene muchas mas llamadas perdidas que los demas esto require mucha atencion.
